# 16. compare on barchat how many married males are there in states: Delhi, Gujarat, Rajasthan, Telangana

```python
df1 =  df[ (df['Whether_married'] == 1) & (df['Gender'] == 'M') & 
          ((df['State'] == 'Delhi')|(df['State'] == 'Gujarat')|(df['State'] == 'Rajasthan')|(df['State'] == 'Telangana'))]  

df2= df1[['Whether_married', 'State']]

df3 =  df[ (df['Gender'] == 'M') & 
          ((df['State'] == 'Delhi')|(df['State'] == 'Gujarat')|(df['State'] == 'Rajasthan')|(df['State'] == 'Telangana'))]                         

# In df2, we have filtered the columns such that it gives the number of married and unmarried males in above mentioned states.
# In df3, we have filtered the columns such that it gives the number of married males in above mentioned states. 


```

```python
grouped = df3.groupby('State')['Whether_married'].sum()
grouped.plot(kind='barh')
plt.show()
# Delhi has the majority married males, while least is seen in Telangana
```
